```python
import numpy as np
import pandas as pd

np.random.seed(42)
n_transactions = 1000000

income_deciles = np.random.randint(1, 11, size=n_transactions)

categories = ['Electronics', 'Clothing', 'Home', 'Sports', 'Toys']
category_choices = np.random.choice(categories, size=n_transactions)

mean_base_price = {
    'Electronics': 100.0,  
    'Clothing':    50.0,
    'Home':        80.0,
    'Sports':      60.0,
    'Toys':        40.0
}
```

```python
std_base_price = {cat: 0.2 * mean for cat, mean in mean_base_price.items()}

base_prices = np.array([
    np.random.normal(mean_base_price[cat], std_base_price[cat]) 
    for cat in category_choices
])
base_prices = np.where(base_prices < 0, 0.1, base_prices)

wtp = base_prices * (1 + np.random.rand(n_transactions) * 0.5)

anchor_flag = np.random.binomial(1, 0.5, n_transactions)  
loss_flag   = np.random.binomial(1, 0.5, n_transactions)  

group_code = anchor_flag + 2 * loss_flag

period = np.concatenate([np.zeros(n_transactions, dtype=int), np.ones(n_transactions, dtype=int)])

income_deciles_all = np.concatenate([income_deciles, income_deciles])
category_all = np.concatenate([category_choices, category_choices])
base_price_all = np.concatenate([base_prices, base_prices])
wtp_all = np.concatenate([wtp, wtp])
group_all = np.concatenate([group_code, group_code])
```

```python
prices = np.empty(2 * n_transactions)

frac_control = 0.25 
frac_anchor_mean = 0.50  
frac_loss_mean = 0.40    
frac_both_mean = 0.70    


decile = income_deciles_all  
anchor_decile_adjust = 1 + (5.5 - decile) / 9 * 0.10   
loss_decile_adjust   = 1 + (5.5 - decile) / 9 * 0.05  
both_decile_adjust   = 1 + (5.5 - decile) / 9 * 0.10   

for i in range(2 * n_transactions):
    if period[i] == 0:
        frac = frac_control  
    else:
        if group_all[i] == 0:
            frac = frac_control
        elif group_all[i] == 1:  
            frac = frac_anchor_mean * anchor_decile_adjust[i]
        elif group_all[i] == 2:  
            frac = frac_loss_mean * loss_decile_adjust[i]
        elif group_all[i] == 3:  
            frac = frac_both_mean * both_decile_adjust[i]
    if frac > 1:
        frac = 1.0
    prices[i] = base_price_all[i] + frac * (wtp_all[i] - base_price_all[i])

data = pd.DataFrame({
    'period': period,  
    'group': np.concatenate([np.repeat('Control', n_transactions),  
                              np.where(group_code==0, 'Control',
                                       np.where(group_code==1, 'Anchoring',
                                                np.where(group_code==2, 'LossAversion', 'Both')))]),
    'income_decile': income_deciles_all,
    'category': category_all,
    'base_price': base_price_all,
    'WTP': wtp_all,
    'price': prices
})

print(data.head())
print("Total records:", len(data))
print("Treatment group counts (period 1):")
print(data[data['period']==1]['group'].value_counts())
```

```python

avg_price_by_group = data[data['period']==1].groupby('group')['price'].mean()
print("Average price in period 1 by group:")
print(avg_price_by_group)
```

```python
df1 = df1.copy()  
X = pd.get_dummies(df1[['income_decile', 'category', 'group']], columns=['category', 'group'], drop_first=False)
y = df1['price']

sample_size = 200000  
train_sample = np.random.choice(df1.index, size=sample_size, replace=False)
X_train = X.loc[train_sample]
y_train = y.loc[train_sample]

rf = RandomForestRegressor(n_estimators=50, max_depth=10, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)

feat_importances = pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending=False)
print("Top 10 feature importances in Random Forest:")
print(feat_importances.head(10))
```

```python
df1.loc[df1['group']!='Control', 'fair_price'] = (
        df1['base_price'] + frac_control * (df1['WTP'] - df1['base_price'])
)
df1.loc[df1['group']=='Control',  'fair_price'] = df1.loc[df1['group']=='Control','price']

df1['consumer_surplus'] = df1['WTP'] - df1['price']
df1['overpayment']      = df1['price'] - df1['fair_price']

welfare_by_decile = df1.groupby('income_decile').agg(
        price             = ('price',            'mean'),
        fair_price        = ('fair_price',       'mean'),
        consumer_surplus  = ('consumer_surplus', 'mean'),
        overpayment       = ('overpayment',      'mean'),
        count             = ('overpayment',      'count')   
)

welfare_by_decile['overpayment_rate'] = None  

PPT = 0.20
for dec in welfare_by_decile.index:
    subset = df1[df1['income_decile'] == dec]
    if len(subset) == 0: 
        continue
    risk = np.mean(subset['price'] > 1.2 * subset['fair_price'])
    welfare_by_decile.loc[dec, 'overpayment_rate'] = risk

print("Income Decile | Avg Price | Avg Fair Price | Avg Consumer Surplus | Avg Overpayment | Overpay >20% (risk)")
for dec, row in welfare_by_decile.iterrows():
    print(f"{dec:13d} | ${row['price']:.2f} | ${row['fair_price']:.2f} | "
          f"${row['consumer_surplus']:.2f} | ${row['overpayment']:.2f} | "
          f"{row['overpayment_rate']*100:.1f}%")
```

```python
groups = ['Control', 'Anchoring', 'LossAversion', 'Both']
colors = ['blue', 'red', 'green', 'purple']

plt.figure(figsize=(8,6))
max_price = df1['price'].max()
bins = np.linspace(0, max_price, 50)

for grp, color in zip(groups, colors):
    prices_grp = df1[df1['group']==grp]['price']
    plt.hist(prices_grp, bins=bins, alpha=0.5, label=grp, color=color)
plt.title("Price Distribution by Treatment Group (Post-Treatment)")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
deciles = sorted(df1['income_decile'].unique())
anchor_effects = []
loss_effects = []

for d in deciles:
    avg_control = df1[(df1['income_decile']==d) & (df1['group']=='Control')]['price'].mean()
    avg_anchor  = df1[(df1['income_decile']==d) & (df1['group']=='Anchoring')]['price'].mean()
    avg_loss    = df1[(df1['income_decile']==d) & (df1['group']=='LossAversion')]['price'].mean()
    anchor_effects.append(avg_anchor - avg_control)
    loss_effects.append(avg_loss - avg_control)

plt.figure(figsize=(8,5))
plt.plot(deciles, anchor_effects, marker='o', label="Anchoring effect")
plt.plot(deciles, loss_effects, marker='s', label="Loss-aversion effect")
plt.axhline(0, color='gray', linestyle='--')
plt.title("Treatment Effect on Price by Income Decile")
plt.xlabel("Income Decile")
plt.ylabel("Price Increase due to Treatment (Difference vs Control)")
plt.legend()
plt.show()
```
